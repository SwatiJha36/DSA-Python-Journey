```python
# reverse of an array

arr = [11,15,23,55]
left = 0
right = len(arr)-1
while left < right:
  arr[left],arr[right]=arr[right],arr[left]
  left += 1
  right -=1
print(f"Reversed array is{arr}")       # Time = 0[n]
                                       # Space = 0[1]
```

```python
#check for Palindrome

s = "Sis".lower()
left = 0
right = len(s)-1
while left < right:
  if s[left]==s[right]:
    left += 1
    right -= 1
    print(f"{s} is a Palindrome")
    break
  else:
    if s[left]!=s[right]:                           # Time = 0[n]
                                                    # Space = 0[1]
      print("False")
      break


```

```python
s = "Madam"
left = 0
right = len(s)-1
while left < right:
  if s[left]!=s[right]:
      print("False")
      break
      left += 1
      right -= 1
  else:
    print("True")
    break




```

```python
# Pair sum for getting exact value not index
arr =[1,2,3,4,6]
target = 7
left = 0
right = len(arr)-1
while left < right:
  total = arr[left] + arr[right]
  if total == target:
    print(arr[left],arr[right])
    break
  elif total<target:                                 # Time = 0[n]
                                                    # Space = 0[1]
    left += 1
  else:
    right -=1
```

```python
# Pair sum for getting index of that value not exact value
arr =[1,2,3,4,6]
target = 7
left = 0
right = len(arr)-1
while left < right:
  total = arr[left] + arr[right]
  if total == target:
    print(left,right)
    break
  elif total<target:
    left += 1
  else:
    right -=1
```

```python
# Pair sum for getting all possible values

arr =[1,2,3,4,6]
target = 7
left = 0
right = len(arr)-1
while left < right:
  total = arr[left] + arr[right]
  if total == target:
    print(arr[left],arr[right])
    left += 1
    right -= 1
  elif total<target:
    left += 1
  else:
    right -=1
```
