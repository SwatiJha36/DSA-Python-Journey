```python
# Prefix array
arr = [6,2,4,1,3]
prefix = [0]*len(arr)
prefix[0]=arr[0]
for i in range(1,len(arr)):
  prefix[i]=prefix[i-1]+arr[i]                  # Time = O(n)
                                                # Space = O(n)
  print("Prefix is ",prefix)
```

```python
# range sum Query
#does only upto the given range and leave others as it is eg. 5,6,9,7,2
# Time = O(n)
# Space = O(1)

arr = [5,1,3,7,2]
for i in range(1,3):
  arr[i]=arr[i]+arr[i-1]
print(arr)
```
