```python
# to find 2nd largest number
arr = [7,2,9,4]

# take large,second,third etc as = to float(-infinity) because it will not take neg numbers
largest = 0
second = 0
for num in arr:
  if num > largest:
   second = largest
   largest = num
  elif num > second and num != largest:
   second = num
print(f"Second largest number is {second}")

```

```python
arr = [7,2,9,4]
largest = 0
second = 0
third = 0
for num in arr:
  if num > largest:
    third = second
    second = largest
    largest = num
  elif num > second and num != largest:
   second = num
  elif num > third and num != largest and num != second:
   third = num
print(f"Three largest number is {largest},{second},{third}")
```

```python
arr = [7,2,9,4]
largest = 0
second = 0
third = 0
fourth = 0
for num in arr:
  if num > largest:
    fourth = third
    third = second
    second = largest
    largest = num
  elif num > second and num != largest:
   second = num
  elif num > third and num != largest and num != second:
   third = num
  elif num > fourth and num != largest and num != second and num != third:
   fourth = num
print(f"Four largest number are {largest},{second},{third},{fourth}")
```
